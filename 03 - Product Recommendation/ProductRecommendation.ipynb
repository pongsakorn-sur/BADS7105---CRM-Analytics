# Market Basket Analysis

```python
df = df.astype('int64')
df.head()
```

```python
df['Transaction'] = df.index
df = df.set_index('Transaction')
df
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(10)
```

```python
rules[ (rules['lift'] >= 1) &
       (rules['confidence'] >= 0.5)&
       (rules['support'] >= 0.3) ].sort_values('confidence', ascending=False)
```

```python
df_rules = rules
df_rules
```

```python
df_rules['ante_count'] = df_rules['antecedents'].apply(lambda x: len(x))
df_rules['conse_count'] = df_rules['consequents'].apply(lambda x: len(x))
df_rules
```

```python
# Selected only one-item pair
df_use = df_rules[(df_rules['ante_count'] <= 1) &
            (df_rules['conse_count'] <= 1) &
            (df_rules['lift'] >= 1.05) &
            (df_rules['confidence'] >= 0.5)&
            (df_rules['support'] >= 0.1) ].sort_values('lift', ascending=False)

df_use
```

```python
import networkx as nx
fig, ax=plt.subplots(figsize=(30,20))
GA=nx.from_pandas_edgelist(df_use,source='antecedents',target='consequents')
nx.draw(GA,with_labels=True)
plt.show()
```

# Collaborative Filtering

```python
item_item_matrix = pd.DataFrame(index=df.columns,columns=df.columns)
item_item_matrix
```

```python
from scipy.spatial.distance import cosine

for i in range(0,len(item_item_matrix.columns)) :
    # Loop through the columns for each column
    for j in range(0,len(item_item_matrix.columns)) :
      # Fill in placeholder with cosine similarities
      item_item_matrix.iloc[i,j] = 1 -cosine(df.iloc[:,i],df.iloc[:,j])
        
item_item_matrix.head()
```

```python
links = item_item_matrix.rename_axis('related item', 
                                     axis='columns').stack().reset_index()
links.columns = ['item', 'related item','value']

links.head()
```

```python
links_filtered=links.loc[ (links['value'] > 0.8) & 
                         (links['item'] != links['related item']) ]

links_filtered.sort_values('value', ascending=False)
```

```python
import networkx as nx
fig, ax=plt.subplots(figsize=(20,15))
GA=nx.from_pandas_edgelist(links_filtered,source='item',target='related item')
nx.draw(GA,with_labels=True)
plt.show()
```
